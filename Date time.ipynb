Objek utama dari module datetime :
- date -> suatu tanggal, tahun, bulan, day
- time -> suatu waktu, hour, minute, second, microsend
- datetime -> suatu tanggal dan waktu, tahun, bulan, day, hour, minute, second, microsend
- timedelta -> 
- tzinfo
- timezone

```python
from datetime import date, time, datetime, timedelta
```

```python
# Cretae Date with date class
mydate = date(year=2020, month=6, day=10)
```

```python
# Extract date attribute
mydate.day
```

```python
# Cretae time with time class
mytime = time(hour=20, minute=30, second=10)
```

```python
# Extract time attribute
mytime.second
```

```python
# Cretae datetime with datetime class
mydatetime = datetime(year=2019, month=11, day=10, hour=10, minute=20)
```

```python
# Extract datetime 
mydatetime.month
```

```python
# Get Current Date
date.today()
```

```python
# Get Current datetime
datetime.now()
```

```python
# Get Current Time
datetime.now().date()
```

```python
# create timedelta
mytimedelta = timedelta(days=10)
```

```python
mydatetime
```

```python
mydatetime + mytimedelta
```

```python
mydatetime2 = datetime(year=2019, month=12, day=10, hour=10, minute=20)
```

```python
mydatetime2 - mydatetime
```

```python
# datetime from string
string1 = '2011-01-20'
string2 = 'Monday, 10 June 1996'
```

```python
type(string1)
```

```python
datetime.strptime(string1, '%Y-%m-%d')
```

```python
datetime.strptime(string2, '%A, %d %B %Y')
```
